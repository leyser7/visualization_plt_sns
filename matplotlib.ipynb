What is Matplotlib?

* Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

* Matplotlib was created by John D. Hunter.

* Matplotlib is open source and we can use it freely.

* Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

Matplotlib Plotting

```python
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

Plotting Without Line

```python
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

Multiple Points

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Nombre del titulo")
plt.xlabel("Label en X")
plt.ylabel("Label en Y")

plt.show()
```

Fonts

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Nombre del titulo", font1)
plt.xlabel("Label en X", font2)
plt.ylabel("Label en Y", font2)

plt.show()
```

```python
plt.plot(x, y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("a la izquierda", font1, loc = 'left')
plt.xlabel("Label en X", font2)
plt.ylabel("Label en Y", font2)

plt.show()
```

### Grid

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Nombre del titulo")
plt.xlabel("Label en X")
plt.ylabel("Label en Y")
plt.grid(axis='x')
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Nombre del titulo")
plt.xlabel("Label en X")
plt.ylabel("Label en Y")
plt.grid(axis='x')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

### Subplots

figure rows, columns and third argument represents the index of the current plot

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

Title

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.title("Titulo 1")
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.title("Titulo 2")
plt.plot(x,y)
plt.suptitle("Un gran titulo")
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

### Histograma

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.ylabel("Numero de personas")
plt.xlabel("Estatura")
plt.show() 
```

```python
plt.hist(x, bins=30)
plt.ylabel("Numero de personas")
plt.xlabel("Estatura")
plt.show() 
```

### Pie charts

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels =mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels =mylabels, startangle = 90)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0, 0, 0.2]
plt.pie(y, labels =mylabels, startangle = 90, explode=myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0, 0, 0.2]
plt.pie(y, labels =mylabels, startangle = 90, explode=myexplode,shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = [0, 0, 0, 0.2]
plt.pie(y, labels =mylabels, startangle = 90, explode=myexplode,shadow = True, colors=mycolors)
plt.legend(title = "Titulo de la legenda:")
plt.show() 
```
